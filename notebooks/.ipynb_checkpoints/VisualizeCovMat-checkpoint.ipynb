```python
def cov2cor(cov):
    Ni, Nj = cov.shape[0], cov.shape[1]
    cor = np.zeros(cov.shape)
    for i in range(Ni):
        for j in range(Nj):
            if(cov[i,i] * cov[j,j] > 0):
                cor[i,j] = cov[i,j] / np.sqrt(cov[i,i]*cov[j,j])
    return cor
```
